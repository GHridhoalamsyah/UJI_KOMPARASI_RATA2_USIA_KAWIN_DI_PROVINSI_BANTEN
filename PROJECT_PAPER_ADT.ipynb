# UJI WILCOXON UNTUK MENGETAHUI APAKAH ADA PERBEDAAN USIA RATA2 KAWIN PERTAMA PADA TAHUN 2019  DAN TAHUN 2023

### DI KABUPATEN

```r
data = read_xlsx("/content/Data ADT.xlsx",sheet = 5)
data
```

```r
data_kabupaten = data
data_kabupaten
```

```r
wilcox.test(data_kabupaten$`2019`, data_kabupaten$`2023`,
            paired = TRUE,
            alternative = "two.sided",  # bisa juga "greater" / "less"
            conf.int = TRUE)
```

p < 0.05 → ada perbedaan signifikan

p ≥ 0.05 → tidak ada perbedaan signifikan


KESIMPULAN = NILAI P VALUE = 0.125 > 0,05 ARTINYA tidak ada perbedaan signifikan rata2 usia kawin di tahun 2019 dan 2023 di kabupaten

### DI KOTA

```r
data = read_xlsx("/content/Data ADT.xlsx", sheet = 6)
data

data_kota = data
```

```r
wilcox.test(data_kota$`2019`, data_kota$`2023`,
            paired = TRUE,
            alternative = "two.sided",  # bisa juga "greater" / "less"
            conf.int = TRUE)
```

p < 0.05 → ada perbedaan signifikan

p ≥ 0.05 → tidak ada perbedaan signifikan


KESIMPULAN = NILAI P VALUE = 0.125 > 0,05 ARTINYA tidak ada perbedaan signifikan rata2 usia kawin di tahun 2019 dan 2023 di kota

### KABUPATEN DAN KOTA

```r
data = read_xlsx("/content/Data ADT.xlsx" , sheet =  4)
data
```

```r
data$`2019` = as.numeric(data$`2019`)
data$`2023` = as.numeric(data$`2023`)
```

# UJI WILCOXON DI UNTUK MENGETAHUI APAKAH ADA PERBEDAAN RATA2 KAWIN PERTAMA DI KABUPATEN DAN KOTA  DARI 2019 SAMPAI 2023

```r
data = read_xlsx("/content/Data ADT.xlsx", sheet = 8)
data
```

```r
data$KABUPATEN = as.numeric(data$KABUPATEN)
data$KOTA = as.numeric(data$KOTA)
```

```r
wilcox.test(data$KABUPATEN,data$KOTA,
            paired = TRUE,
            alternative = "two.sided",  # bisa juga "greater" / "less"
            conf.int = TRUE)
```

```r
wilcox.test(data$KABUPATEN, data$KOTA,
            paired = TRUE,
            exact = FALSE,
            conf.int = TRUE)
```

p < 0.05 → ada perbedaan signifikan

p ≥ 0.05 → tidak ada perbedaan signifikan


KESIMPULAN = NILAI P VALUE = 0.0007265 < 0,05

ARTINYA  ada perbedaan signifikan rata2 usia kawin di tahun 2019 SAMPAI 2023 di kabupaten dan kota
